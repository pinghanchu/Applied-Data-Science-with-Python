```python
import pandas as pd
census_df = pd.read_csv('census.csv')
df_copy = census_df[census_df['SUMLEV'] == 50]
df_copy = df_copy.set_index('STNAME')
df_copy = df_copy.drop('District of Columbia')
df_state = df_copy.index.unique()
df_ser = pd.Series()
for i in range(len(df_state)):  
    df_AL = df_copy.loc[df_state[i],['CTYNAME','CENSUS2010POP']].sort_values(['CENSUS2010POP','CTYNAME'], ascending=[False, False])
    sum = 0
    for j in range(3):
        sum = sum+df_AL.iloc[j][1]
    df_ser = df_ser.append(pd.Series({df_state[i]:sum}))
df_pop = pd.DataFrame(df_ser, columns = ['POP'])
print(df_pop.sort_values('POP',ascending=[False])[:3].index.tolist())
```
